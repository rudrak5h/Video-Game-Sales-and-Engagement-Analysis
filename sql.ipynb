```python
import pandas as pd
import sqlite3
```

```python
games = pd.read_csv("cleaned_games.csv")

sales = pd.read_csv("cleaned_vgsales.csv")

merged = pd.read_csv("merged_games_sales.csv")
```

```python
conn = sqlite3.connect("videogame.db")

cursor = conn.cursor()

print("Database connected successfully")
```

```python
games.to_sql(
    'games',
    conn,
    if_exists='replace',
    index=False
)

sales.to_sql(
    'sales',
    conn,
    if_exists='replace',
    index=False
)

merged.to_sql(
    'merged_games_sales',
    conn,
    if_exists='replace',
    index=False
)

print("Tables created successfully")
```

```python
query = """
SELECT name
FROM sqlite_master
WHERE type='table';
"""

pd.read_sql(query, conn)
```

SQL QUERY

```python
#1 Top Rated Games

query = """
SELECT Title, Rating
FROM games
ORDER BY Rating DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#2 Best Selling Platforms

query = """
SELECT Platform,
SUM(Global_Sales) AS Total_Sales
FROM sales
GROUP BY Platform
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#3 Top Publishers

query = """
SELECT Publisher,
SUM(Global_Sales) AS Total_Sales
FROM sales
GROUP BY Publisher
ORDER BY Total_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#4 Most Wishlisted Games

query = """
SELECT Title, Wishlist
FROM games
ORDER BY Wishlist DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#5 Genre with Highest Global Sales

query = """
SELECT Genre,
SUM(Global_Sales) AS Total_Sales
FROM merged_games_sales
GROUP BY Genre
ORDER BY Total_Sales DESC;
"""

pd.read_sql(query, conn)
```

```python
#6 Platform with Most High Rated Games

query = """
SELECT Platform,
COUNT(*) AS High_Rated_Games
FROM merged_games_sales
WHERE Rating >= 4
GROUP BY Platform
ORDER BY High_Rated_Games DESC;
"""

pd.read_sql(query, conn)
```

```python
#7 Regional Sales Comparison

query = """
SELECT
SUM(NA_Sales) AS North_America,
SUM(EU_Sales) AS Europe,
SUM(JP_Sales) AS Japan,
SUM(Other_Sales) AS Other_Regions
FROM sales;
"""

pd.read_sql(query, conn)
```

```python
#8 Average Rating Per Genre

query = """
SELECT Genres,
AVG(Rating) AS Average_Rating
FROM games
GROUP BY Genres
ORDER BY Average_Rating DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#9 Top 10 Global Best Sellers

query = """
SELECT Name,
Global_Sales
FROM sales
ORDER BY Global_Sales DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
#10 Wishlist vs Sales Analysis

query = """
SELECT Title,
Wishlist,
Global_Sales
FROM merged_games_sales
ORDER BY Wishlist DESC
LIMIT 10;
"""

pd.read_sql(query, conn)
```

```python
conn.close()

print("Database connection closed")
```
