```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option('display.max_columns', None)

print("Libraries Imported Successfully")
```

```python
games = pd.read_csv("games.csv")
sales = pd.read_csv("vgsales.csv")
```

```python
print("Games Shape:", games.shape)
print("Sales Shape:", sales.shape)
```

```python
games.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
games.isnull().sum()
```

```python
sales.isnull().sum()
```

```python
games['Rating'].fillna(games['Rating'].median(), inplace=True)

games['Team'].fillna('Unknown', inplace=True)

games['Summary'].fillna('No Summary', inplace=True)
```

```python
sales['Year'].fillna(sales['Year'].median(), inplace=True)

sales['Publisher'].fillna('Unknown', inplace=True)
```

```python
def convert_k(value):

    if isinstance(value, str):

        value = value.replace(',', '').strip()

        if 'K' in value:
            return float(value.replace('K', '')) * 1000

        elif 'M' in value:
            return float(value.replace('M', '')) * 1000000

        elif value == '':
            return np.nan

        else:
            try:
                return float(value)
            except:
                return np.nan

    return value
```

```python
columns_to_convert = [
    'Times Listed',
    'Number of Reviews',
    'Plays',
    'Playing',
    'Backlogs',
    'Wishlist'
]

for col in columns_to_convert:
    games[col] = games[col].apply(convert_k)
```

```python
games.info()
```

```python
games['Release Date'] = pd.to_datetime(
    games['Release Date'],
    errors='coerce'
)
```

```python
games['Release Date'].isnull().sum()
```

```python
games['Release Date'].fillna(
    pd.Timestamp('2000-01-01'),
    inplace=True
)
```

```python
games['Release_Year'] = games['Release Date'].dt.year
```

```python
games['Title'] = games['Title'].str.strip()

sales['Name'] = sales['Name'].str.strip()

games['Genres'] = games['Genres'].str.strip()

sales['Genre'] = sales['Genre'].str.strip()

sales['Publisher'] = sales['Publisher'].str.strip()
```

```python
games.isnull().sum()
```

```python
sales.isnull().sum()
```

```python
games.to_csv("cleaned_games.csv", index=False)

sales.to_csv("cleaned_vgsales.csv", index=False)

print("Cleaned datasets saved successfully")
```

```python
merged = pd.merge(
    games,
    sales,
    left_on='Title',
    right_on='Name',
    how='inner'
)
```
